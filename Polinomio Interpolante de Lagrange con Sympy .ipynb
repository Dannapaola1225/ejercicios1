# ***Polinomio Interpolante de Lagrange con Sympy***

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import symbols, simplify, lambdify
```

```python
 #Calculamos el polinomio de interpolación de Lagrange
def lagrange_interpolacion(x, y):
    n = len(x)
    x_sim = sp.symbols('x')
    polinomio = 0

    for i in range(n):
        termino = y[i]
        for j in range(n):
            if i != j:
                termino *= (x_sim - x[j]) / (x[i] - x[j])
        polinomio += termino

    return sp.expand(polinomio)
```

```python
#evaluamos el polinomio
def evaluar_polinomio(polinomio, valor):

    x_sim = sp.symbols('x')
    return float(polinomio.subs(x_sim, valor))
```

```python
# Función para graficar los puntos y el polinomio interpolante
def graficar_lagrange(x, y, polinomio):
  # Creamos los puntos para la gráfica
    x_vals = np.linspace(min(x) - 1, max(x) + 1, 500)
    y_vals = [evaluar_polinomio(polinomio, val) for val in x_vals]
  # En este espacio, graficamos la función
    plt.scatter(x, y, color='red', label='Puntos dados')
    plt.plot(x_vals, y_vals, label='Polinomio de Lagrange', color='blue')
    plt.grid()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.title('Interpolación de Lagrange')
    plt.show()
```

```python
if __name__ == "__main__":
    # Entrada de datos
    x = [float(i) for i in input("Ingrese los valores de x separados por comas: ").split(",")]
    y = [float(i) for i in input("Ingrese los valores de f(x) separados por comas: ").split(",")]

    # Calcular el polinomio de Lagrange
    polinomio = lagrange_interpolacion(x, y)

    # Mostrar el polinomio
    print("\nEl polinomio de Lagrange es:")
    print(polinomio)

    # Evaluar el polinomio en un punto dado
    xi = float(input("\nIngrese el valor de x para evaluar el polinomio: "))
    resultado = evaluar_polinomio(polinomio, xi)
    print(f"\nEl polinomio evaluado en x = {xi} es: {resultado:.5f}")

    # Graficar el polinomio y los puntos
    graficar_lagrange(x, y, polinomio)
```
