# Gráfica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return (x+1)**(1/2)
def g(x):
    return 1+x*(1/2)
def h(x):
    return 1+x*(1/2)-(x**2)*(1/8)
def d(x):
    return 1+x*(1/2)-(x**2)*(1/8)+(x**3)*(1/16)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
plt.axhline(0, color="pink")
plt.axvline(0, color="pink")
plt.xlim(-10, 10)
plt.ylim(-10, 15)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
