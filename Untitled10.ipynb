# ***Método de Horner***
programa en el que se implementa el método de Newton para polinomios, realizando la evaluación del polinomio y de su derivada mediante el método de Horner.
P(x)=$2x^{4}-3x^{2}+ 3x +*4*$


```python
from math import *
import math
import numpy as np
import matplotlib . pyplot as plt
```

```python
def f(x):
    return 2*x**4 - 3*x**2 + 3*x + 4
def df(x):
    return 8*x**3 - 6*x + 3
```

```python
# Generar valores para x
x = np.linspace(-10, 10, 200)

# Crear la gráfica
plt.plot(x, f(x), color="purple", label='f(x) = $2x^{4}- 3x^{2} + 3x + 4$')  # Gráfica de f(x)


plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title('Gráfica de la función f(x) ')#titulo de la gráfica
plt.legend()#imprime leyenda
plt.grid(color="pink")#imprime cuadricula color rosa
plt.xlabel('Eje X')# enunciado del eje x
plt.ylabel('Eje Y')# enunciado del eje y

plt.xlim(-5, 10)
plt.ylim(0, 20)
#muestra la gráfica
plt.show()
```
