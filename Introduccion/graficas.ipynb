# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
def d(x):
    return x
def h(x):
    return -3**(-x)
```

```python
f(-5),f(1)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x), color="pink")
plt.plot(x, d(x), color="purple")
plt.plot(x, h(x), color="green")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title("suma de funciones")
#plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.

# Mostrarlo.
plt.show()
```
