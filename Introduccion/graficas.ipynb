# ejemplo polinomio de Maclaurin de grado tres $f(x)=e^{x/2} sin^{x/3}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return np.exp(x/2)*np.sin(x/3)
def g(x):
    return (1/3)*x
def h(x):
    return (1/3)*x+(1/6)*x**(2)
def d(x):
    return (1/3)*x+(1/6)*x**(2)+(23/648)*x**3
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
plt.title("polinomios de maclauren")
#plt.axhline(0, color="black")
#plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```
