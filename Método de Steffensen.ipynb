```python
from math import *
import math
import numpy as np
import matplotlib . pyplot as plt # libreria para graficar
```

```python
#definimos la función con la que vamos a trabajar
def f(x):
    return  np.cos(x)-2*(x)+3 # función original
def g(x):
    return  (np.cos(x)+3)/2 # función transformada de f(x) => 0(x)
```

```python
# Generar valores para x
x = np.linspace(-20, 20, 400)

# Crear la gráfica
plt.plot(x, f(x),'b.-', color="blue", label='f(x) = cos(x) - 2x + 3')  # Gráfica de f(x)
plt.plot(x, g(x), color="purple", label='g(x) = (cos(x) + 3) / 2')  # Gráfica de g(x)

plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title('Gráfica de las funciones ')#titulo de la gráfica
plt.legend()#imprime leyenda
plt.grid(color="pink")#imprime cuadricula color rosa
plt.xlabel('Eje X')# enunciado del eje x
plt.ylabel('Eje Y')# enunciado del eje y
# anotacion de donde se intersectan las funciones
plt.annotate('Intersección aproximada', xy=(0.9, 1.795), xytext=(2, 2),
             arrowprops=dict(facecolor='green', shrink=0.005))
# Coordenadas de la raiz de f(x)
x = [1.523586]
y = [1.523597]
#anotacion de donde se encuentra la raiz
plt.annotate('Raiz de la función', xy=(1.523586, 1.523586), xytext=(2, 1.6),
             arrowprops=dict(facecolor='red', shrink=0.005))
plt.scatter(x, y, color='red')

plt.xlim(0, 3)
plt.ylim(0, 3)
#muestra la gráfica
plt.show()
```

```python
#guarda valores iniciales del error, número de iteraciones y la función
fun= f(x)
p0=1 # aproximación inicial
Nom = 100 #número maximo de iteraciones
tol = 10**(-5) #tolerancia de 0.00001
error =100
niter = 0 #contador de iteraciones (i)
```

```python
#estructura de la tabla
print("# iter\t\t p0 \t\t pn+1 \t\tp \t\t error")#imprimimos el encabezado
# comenzamos a aplicar el método de Steffensen
#realizamos el ciclo de iteraciones
for i in range( Nom + 1) : #ciclo iterativo
       # Calcular p1 y p2
        p1 = g(p0)
        p2 = g(p1)

        # Aplicar la fórmula de Steffensen
        if niter == 0:
           d = 0  # Si es la primera iteración, p es cero
        elif (niter<Nom):
           d = p0 - ((p1 - p0) ** 2) / (p2 - 2 * p1 + p0)
        else:
          d=0

        #calculamos el error de la iteración
        if niter == 0:
           error = 0  # Si es la primera iteración, el error es cero
        else:
           error=abs(d-p0)

#imprime los valores arrojados
        print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t{4:6.6f} ".format(niter, p0, p1, d, error ))

#ponemos una condición en dónde si el error es menor que la tolerancia, el procedimiento será exitoso
        if error < tol:
         print(f"Convergencia alcanzada en la iteración {niter}.")
        break  # Salimos del bucle si se alcanza la tolerancia

        p0=p1#damos el nuevo valor a p0 que seria p1, esto es para la sig. iteración
        niter += 1 #aumenta el numero del contador de iteraciones


print("La raíz de la función es {0:6.7f}".format(d))
print ("con un error de:{0:6.7f}".format(error))
```
