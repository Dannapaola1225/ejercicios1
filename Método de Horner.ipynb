# ***Método de Horner***
programa en el que se implementa el método de Newton para polinomios, realizando la evaluación del polinomio y de su derivada mediante el método de Horner.
P(x)=$2x^{4}-3x^{2}+ 3x +4$


```python
from math import *
import math
import numpy as np #para generar valores de x
import matplotlib . pyplot as plt #para fraficar
import sympy as sp #para derivar
```

```python
def f(x):
    return 2*x**4 - 3*x**2 + 3*x - 4
def d_f(x):
    return 8*x**3 - 6*x + 3
```

```python
# Generar valores para x
x = np.linspace(-10, 10, 200)

# Crear la gráfica
plt.plot(x, f(x), color="purple", label='f(x) = $2x^{4}- 3x^{2} + 3x + 4$')  # Gráfica de f(x)
plt.plot(x, d_f(x), color="red", label='d_f(x) = $8x^{3}- 6x + 3$')  # Gráfica de d_f(x)

plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.title('Gráfica de la función f(x) ')#titulo de la gráfica
plt.legend()#imprime leyenda
plt.grid(color="pink")#imprime cuadricula color rosa
plt.xlabel('Eje X')# enunciado del eje x
plt.ylabel('Eje Y')# enunciado del eje y
# Coordenadas de la raiz de f(x)
x = [-1.738956]
y = [0]
#anotacion de donde se encuentra la raiz
plt.annotate('Raiz de la función', xy=(-1.738956, 0), xytext=(2, 2.6),
             arrowprops=dict(facecolor='red', shrink=-1.55))
plt.scatter(x, y, color='red')
plt.xlim(-5, 10)
plt.ylim(-10, 20)
#muestra la gráfica
plt.show()
```

```python
#guardamos los valores iniciales
X0 = -2 # el valor inicial
Nom = 100 #número maximo de iteraciones
tol = 10**(-5) #tolerancia de 0.00001
niter = 0 #contador de iteraciones (i)
error = 100 #error inicial
```

```python
#Método de Horner

#estructura de la tabla
print("# iter\t\t Xn \t\t p(X0) \t\t p´(X0) \t\t Xn_1 \t\t error")

#definimos valores
Xn = X0
def_pol = f(Xn)
def_der = d_f(Xn)

#realizamos el ciclo de iteraciones
Xn_1= Xn-(def_pol / def_der)#aquí se calcula p para las iteraciones
error=abs(Xn_1-Xn)#calculamos el error de la iteración

#imprime los valores arrojados
print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t {4:6.6f} \t {5:6.6f}".format(niter, Xn, def_pol, def_der, Xn_1, error))

# Expresamos la condición
while error > tol and niter < Nom:  # ciclo iterativo
    niter += 1  # aumenta el número del contador de iteraciones
    Xn = Xn_1  # guarda el valor de Xn_1 en Xn
    def_pol = f(Xn)
    def_der = d_f(Xn)

    Xn_1 = Xn - (def_pol / def_der)  # aquí se calcula p para las iteraciones
    error = abs(Xn_1 - Xn)  # calculamos el error de la iteración


    print("{0} \t\t {1:6.6f} \t {2:6.6f} \t {3:6.6f} \t{4:6.6} \t{5:6.6}".format(niter, Xn, def_pol, def_der, Xn_1, error)) #imprime los valores arrojados
    if error <= tol:
     break
print("La raíz de la función dada es {0:6.7f}".format(Xn))
print ("con un error de:{0:6.7f}".format(error))
```
